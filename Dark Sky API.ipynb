```python
from dotenv import load_dotenv
load_dotenv()
import os

API_KEY = os.getenv("DARKSKY_API_KEY")
```

```python
# Using the URL on http://darksky.net/dev/account
# print out the current temperature 

import requests
response = requests.get(f'https://api.darksky.net/forecast/{API_KEY}/37.8267,-122.4233')
ds = response.json()
print(ds.keys())
print(ds['currently'])
print(ds['currently']['temperature'])
```
